## 2. Feature Selection - Correlation

*  Remove all the features which are highly correlated

```python
#imports
from sklearn.datasets import load_boston
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#loading the dataset
data = load_boston()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['Target'] = data.target
```

```python
df.shape
```

```python
data.feature_names
```

```python
len(data.target)
```

```python
df.head()
```

```python
#get independent features
X = df.drop("Target", axis=1)
#get dependent features
y = df['Target']
```

```python
X.head()
```

```python
y.head()
```

```python
#seperate dataset into train and test data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=0)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
X_train.corr()
```

**Plot using pearson correlation**

```python
import seaborn as sns
plt.figure(figsize=(12,10))
cor=X_train.corr()
sns.heatmap(cor, annot=True, cmap=plt.cm.CMRmap_r)
plt.show()
```

**function to select highly correlated features**

```python
#remove the first feature that is correlated with other features
def correlation(dataset, threshold):
    col_corr = set()
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j])>threshold:
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
                
    
    return col_corr      
        
```

```python
#correlation features of X_train dataset with threshold=0.7
corr_features = correlation(X_train, 0.7)
len(corr_features)
```

```python
corr_features
```

```python
#drop the correlated features from training dataset
X_train.drop(corr_features, axis=1)
```

```python
#drop the correlated features from test dataset
X_test.drop(corr_features, axis=1)
```
